**Installer k9s**

K9s est un outil open-source en ligne de commande qui permet aux utilisateurs de gérer et de surveiller des clusters Kubernetes de manière plus efficace. Il fournit une interface en ligne de commande intuitive et facile à utiliser pour les administrateurs système et les développeurs pour interagir avec les ressources Kubernetes, notamment les nœuds, les déploiements, les services, les pods et les volumes.

K9s fournit des informations détaillées sur l'état des ressources Kubernetes, telles que les données de surveillance des conteneurs, les logs, les événements et les métriques. Il permet également aux utilisateurs de filtrer, trier et rechercher des ressources en utilisant divers critères.

https://k9scli.io/topics/install/

## II. Outils optionnels qui peuvent être utiles

#### Installer le prompt kube-ps1

Il s'agit d'une extension de prompt pour la ligne de commande qui fournit des informations utiles sur votre contexte Kubernetes actuel, ce qui peut être très pratique lorsque vous travaillez avec des clusters Kubernetes. 

- Informations sur le contexte Kubernetes : le prompt kube-ps1 affiche le nom du cluster, le namespace et le contexte actuel de votre cluster Kubernetes, ce qui peut vous aider à éviter les erreurs de configuration en vous assurant que vous travaillez dans le bon contexte.

- Visibilité des erreurs : lorsque vous travaillez avec plusieurs clusters Kubernetes, il peut être facile de se tromper et de travailler sur le mauvais cluster. Le prompt kube-ps1 peut vous aider à éviter cela en affichant un message d'erreur si vous essayez d'exécuter une commande dans le mauvais contexte.

- Productivité améliorée : en affichant des informations utiles sur le contexte Kubernetes, le prompt kube-ps1 peut vous aider à travailler plus rapidement et plus efficacement en vous permettant de visualiser rapidement votre environnement de développement.

https://github.com/jonmosco/kube-ps1

**Installer Tilt**

Tilt est un outil open-source pour Kubernetes qui vise à améliorer le flux de travail de développement en permettant une itération rapide et une boucle de rétroaction rapide pour les développeurs.

Tilt simplifie le processus de développement en permettant aux développeurs de travailler sur leur code et de voir les changements en temps réel dans un environnement de développement local. Tilt utilise des conteneurs Docker pour exécuter des applications localement et fournit un ensemble de fonctionnalités pour faciliter la construction et le déploiement de ces conteneurs.

En utilisant Tilt, les développeurs peuvent éviter de devoir construire et déployer manuellement leur application à chaque modification de code. Au lieu de cela, Tilt surveille les fichiers du projet pour détecter les modifications et reconstruit et redéploie automatiquement l'application lorsqu'un changement est détecté.

Tilt fournit également une interface utilisateur pour visualiser l'état des applications, accéder aux logs et débuger les conteneurs en cours d'exécution. Il prend également en charge le débogage en direct avec des points d'arrêt et des instructions de débogage.

https://docs.tilt.dev/install.html

## III. Création d'un cluster Kube

Il existe de nombreuses façon de créér un cluster Kube aujourd'hui (CoLiMa, KinD, k3d, MicroK8s, MiniKube, Rancher Desktop, etc.). Nous utiliserons aujourd'hui Docker Desktop.

Les avantages : 
- simplicité
- linux, windows, mac
- interface graphique
- Facile à installer
- bonne intégration avec l'environnement docker (ex: docker build va rendre dispo automatiquement cette image sur Kubernetes)

Les inconvénients :
- gratuit mais uniquement pour une utilisation personnelle
- limité : ça donne juste un cluster avec un noeud, et on ne peut pas choisir la version de Kubernetes

```shell
kubectl config view
```

Vous devriez avoir une config vide car le cluster n'est pas encore lancé :
```
apiVersion: v1
clusters: null
contexts: null
current-context: ""
kind: Config
preferences: {}
users: null
```

#### Créér un cluster Kube

Aller sur le dashboard de Docker Desktop, puis dans les Préférences, Kubernetes, checké la case "Enable Kubernetes", puis "Apply & Restart"
